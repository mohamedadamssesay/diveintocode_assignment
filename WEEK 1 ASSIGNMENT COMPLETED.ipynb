# [Problem 1] Create using the exponentiation arithmetic operator

```python
Thick=0.00008

Thick*=2**43
print("Folded_paper_Thickness in meters ={}m".format(Thick))
```

# [Problem 2] Unit Conversion

```python
Thick=0.00008

Thick*=2**43
print("Folded_paper_Thickness in meters ={:.2f}Km".format(Thick/1000))
```

# [Problem 3]  Create Using a for statement

```python
thick=0.00008
for i in range(43):
    thick=thick*2
print("Folded_paper_Thickness in Kilometers ={} Km".format(thick/1000))
```

# [Problem 4] Comparision of Calculation time

```python
%%timeit
thick=0.00008
for i in range(43):
    thick=thick*2
```

# [Problem 5 ] Saving to list

```python
thick=0.00008
list =[]
for fold in range (1,45):
    list.append(thick*2**fold)
len(list)
```

# [Problem 6] Displaying a line Graph

```python
import  matplotlib.pyplot  as plt
%matplotlib inline
plt.title( " thick of folded paper " )
plt.xlabel( " number of folds " )
plt. ylabel( " thick [m] " )
plt.plot(list) # put list variable name in place of "list name" 
plt.show()
```

# [Problem 7] Customizing graphs

```python
import matplotlib
import seaborn as sns
sns.set_style("darkgrid")
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.plot(list,color='red')
plt.title('Thickness of Folding paper')
plt.xlabel('Folding')
plt.ylabel('Thick [m]')
plt.show()
```

```python
values = range(1,45)
plt.plot(values,list,'b-x',linewidth=4,markersize=12,markeredgewidth=4,markeredgecolor='navy')
plt.title('Thickness of Folded paper')
plt.xlabel('Folding')
plt.ylabel('Thickness')
plt.show()
```

```python
values = range(1,45)
plt.plot(values,list,'r-.',linewidth=4,markersize=12,markeredgewidth=4,markeredgecolor='navy')
plt.title('Thickness of Folded paper')
plt.xlabel('Folding')
plt.ylabel('Thickness')
plt.show()
```
